```python
import numpy as np
import pandas as pd
```

```python
#read the data 
data=pd.read_csv("D:\Beinex\Python\Class learning\Pandas\pandas -data\Sample - Superstore.xls - Orders.csv")
data.head()
```

```python
#datatypes
data.info()
```

```python
#coverting the date into date format
data['Order Date']=pd.to_datetime(data['Order Date'])
#parse the year 
data['order_year']=data['Order Date'].dt.year
```

```python
data['order_year']
```

```python
#Aggregate the data to the years each customer made an order
data_1=data.groupby(['Customer ID','order_year'])['Order ID'].count().reset_index(name='Order_count')
data_1.head(20)
```

```python
#Calculate the year each customer made their First Purchase
data_2=data_1.groupby('Customer ID')['order_year'].min().reset_index(name='First_purchase')
data_2
```

```python
#merging the two tables
data_f=pd.merge(data_1,data_2,left_on=['Customer ID'],right_on=['Customer ID'])
data_f
```

```python
#creating unique customer IDS
customer_ID=data_f['Customer ID'].unique()
#extracting the years of purchasing from data
years=range(data_f['order_year'].min(),data_f['order_year'].max()+1)
```

```python
#combining both s.t a combination of unique customer IDs and years of order are made
import itertools
data_l=list(itertools.product(customer_ID,years))
data_l=pd.DataFrame(data_l,columns=['Customer ID','year_of_order'])
data_l
```

```python
#commbined the data_l table and data_f table into data_n
data_n=pd.merge(data_l,data_f,how='left',left_on=['Customer ID','year_of_order'],right_on=['Customer ID','order_year'])
```

```python
data_n.head(20)
```

```python
#Create a field to flag these new rows, making it clear whether a customer placed an order in that year or not
data_n['Purchased']=np.where(data_n['Order_count']>0,1,0)
data_n.head(30)
```

```python
data_n['cust_2018'] = " "
data_n['cust_2019'] = " "
data_n['cust_2020'] = " "
data_n['cust_2021'] = " "

for i, row in data_n.iterrows():
    if row['year_of_order']==2018 and row['Purchased']==1:
        data_n.loc[i, 'cust_2018']="yes"
    elif row['year_of_order'] == 2019 and row['Purchased']==1:
        data_n.loc[i, 'cust_2019']="yes"
    elif row['year_of_order']==2020 and row['Purchased']==1:
        data_n.loc[i, 'cust_2020']="yes"
    elif row['year_of_order']==2021 and row['Purchased']==1:
        data_n.loc[i, 'cust_2021'] = "yes"

data_n.head(30)
```

```python
data_n['cust_2018']=data_n.groupby('Customer ID')['cust_2018'].transform(lambda x: x.str.cat(sep=''))
data_n['cust_2019']=data_n.groupby('Customer ID')['cust_2019'].transform(lambda x: x.str.cat(sep=''))
data_n['cust_2020']=data_n.groupby('Customer ID')['cust_2020'].transform(lambda x: x.str.cat(sep=''))
data_n['cust_2021']=data_n.groupby('Customer ID')['cust_2021'].transform(lambda x: x.str.cat(sep=''))
data_n.head(30)
```

```python
data_n = pd.DataFrame(data_n.drop_duplicates(subset='Customer ID', keep='first'))
data_n
```

```python
data_n.info()
```

```python
data_n['cust_2018'] = data_n['cust_2018'].apply(lambda x: 1 if 'yes' in x else 0)
data_n['cust_2019'] = data_n['cust_2019'].apply(lambda x: 1 if 'yes' in x else 0)
data_n['cust_2020'] = data_n['cust_2020'].apply(lambda x: 1 if 'yes' in x else 0)
data_n['cust_2021'] = data_n['cust_2021'].apply(lambda x: 1 if 'yes' in x else 0)
data_n
```

```python
data_n['cust_type']='other'

for i, row in data_n.iterrows():
    if  row['cust_2020']==1 and row['cust_2021']==1:
        data_n.loc[i, 'cust_type']="Consistent"
    elif row['cust_2020']==1 and row['cust_2021']==0:
        data_n.loc[i, 'cust_type']="Sleeping"  
    elif row['cust_2018']==1|row['cust_2019']==1 and row['cust_2020']==0 and row['cust_2021']==0:
        data_n.loc[i, 'cust_type']="Sleeping"  
    elif row['cust_2019']==1 and row['cust_2020']==0 and row['cust_2021']==0:
        data_n.loc[i, 'cust_type']="Sleeping" 
    elif row['cust_2020']==0 and row['cust_2021']==1:
        data_n.loc[i, 'cust_type']="Returning "
    elif row['cust_2018']==0 and row['cust_2019']==0 and row['cust_2020']==0 and row['cust_2021']==1:
        data_n.loc[i, 'cust_type']="New"
data_n.head(150)
```
