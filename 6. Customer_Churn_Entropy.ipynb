```python
#normalization of DayMins coloumn in data
from sklearn.preprocessing import MinMaxScaler
data[['DayMins','DayCalls']]=MinMaxScaler().fit_transform(data[['DayMins','DayCalls']])

data
```

```python
#selecting the predictors and label value
X=data[["AccountWeeks","ContractRenewal","DataPlan","DataUsage","CustServCalls","DayMins","DayCalls","MonthlyCharge","OverageFee","RoamMins"]]
Y=data['Churn']
```

```python
#training and testing at 70:30 ratio
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.7,test_size=0.3,random_state=1)
```

```python
#information gain condition for feature selection 
#Decision tree based prediction where the criterion is set to 'entropy'
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score

dt=DecisionTreeClassifier(max_depth=6,random_state=1,criterion='entropy')
dt.fit(X_train,Y_train)
Y_Pred=dt.predict(X_test)
acc=accuracy_score(Y_test,Y_Pred)
print("Test accuracy: ",acc)
```

```python
#training and testing at 80:20 ratio
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,test_size=0.2,random_state=1)

dt=DecisionTreeClassifier(max_depth=6,random_state=1,criterion='entropy')
dt.fit(X_train,Y_train)
Y_Pred=dt.predict(X_test)
acc=accuracy_score(Y_test,Y_Pred)
print("Test accuracy: ",acc)
```

```python
#training and testing at 90:10 ratio
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.9,test_size=0.1,random_state=1)

dt=DecisionTreeClassifier(max_depth=6,random_state=1,criterion='entropy')
dt.fit(X_train,Y_train)
Y_Pred=dt.predict(X_test)
acc=accuracy_score(Y_test,Y_Pred)
print("Test accuracy: ",acc)
```

```python
#The best accuracy achieved with 80:20 split
#training and testing at 80:20 ratio
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,test_size=0.2,random_state=1)

dt=DecisionTreeClassifier(max_depth=6,random_state=1,criterion='entropy')
dt.fit(X_train,Y_train)
Y_Pred=dt.predict(X_test)
acc=accuracy_score(Y_test,Y_Pred)
print("Test accuracy: ",acc)
```

```python
fig=plt.figure(figsize=(25,20))
_=tree.plot_tree(dt,filled=True)
```

From the decision tree it is found that CustServCalls is the most influencing factor in determining the customer churn.This is followed by 'MonthlyCharge' and 'DayMins'

```python
def check_prime(x):
    if x<=1:
        return False
    elif x==2 or x==3:
        return True
    elif x>2:
        for i in range (2,int(x**0.5+1)):
            if x%i==0:
                return False
            else:
                return True
```
