1. NLP using **nltk** package

```python
import pandas as pd
import matplotlib.pyplot as plt

import nltk
nltk.download('punkt')
```

```python
data=open(r"C:\Users\ginus\OneDrive\Desktop\Howard_Garis_adventures.txt").read()
print(data)
```

```python
len(data)
```

```python
#Applying tokenization
from nltk.tokenize import word_tokenize,sent_tokenize
data_wtok= word_tokenize(data)
print(data_wtok)

#data_stok= sent_tokenize(data)
#print(data_stok)
```

```python
#Tried using re library to find regular expressions
import re
data_r=re.findall(r'\w+',data)
print(data_r)
len(data_r)
```

```python
#converting all tokens to lowercase
data_l=[token.lower() for token in data_r]
```

```python
#Continues using  data data_l
#remove stopword from the text
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words=stopwords.words("english")


def rem_stopwords(text):
    data_filtered = []
    for word in text:
        if word not in stop_words:
            data_filtered.append(word)
    return data_filtered

data_s=rem_stopwords(data_l)
```

```python
len(data_s)
```

*We can find a considerable decrease in the number of words for analysis*

```python
#Applying lemmatization of data
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer as lm
data_lem=[lm().lemmatize(word) for word in data_s]
```

```python
#Applying stemming of data
from nltk.stem import PorterStemmer as ps
data_st=[ps().stem(word) for word in data_lem]
len(data_st)
```

```python
#tried to apply n gram 
from nltk import ngrams
n=10
grams=ngrams(data_l,n)
for i in grams:
    print(i)
```

3. Doing NLP pre-processing without using any library

```python
data=open(r"C:\Users\ginus\OneDrive\Desktop\Howard_Garis_adventures.txt").read()
print(data)
```

```python
#convert the data to lowercase
data=data.lower()
```

```python
punctuations = "!\"#$%&'()*+,-./:;<=>?@[\\]^_`{|}~"

data_wpt = ""
for char in data:
    if char not in punctuations:
        data_wpt += char
```

```python
#applying tokenization
data_tok=data_wpt.split()
```

```python
#create a list of stopwrods
stpwrds=set(["a", "an", "the", "is", "and", "or", "not","our","they","them","she","he","you","some","so","did"])

def stopwords_rem(text):
    data_filtered = []
    for word in text:
        if word not in stpwrds:
            data_filtered.append(word)
    return data_filtered

data_ws=stopwords_rem(data_tok)
```

```python
len(data_ws)
```

```python
#applying stemming of data without any library feels like a complex process. a simple code 
# of how the process can be done from scratch is given below 
def stem(word):
    if word.endswith("sses"):
        return word[:-4] + "ss"
    elif word.endswith("ies"):
        return word[:-3] + "i"
    elif word.endswith("ss"):
        return word
    elif word.endswith("s"):
        return word[:-1]
    elif word.endswith("ing"):
        return word[:-3]
    return word

tokens = ["running", "jumps", "studies", "happily", "cats"]

stems = [stem(token) for token in tokens]
print("Stems:", stems)
```
