```python
#coverting the date into date format
data['Order Date']=pd.to_datetime(data['Order Date'],format='%d/%m/%Y')
```

```python
#parse the year 
data['order_year']=data['Order Date'].dt.year
```

```python
data['order_year']
```

```python
#Calculate the year each customer made their First Purchase
data_2=data.groupby('Customer ID')['order_year'].min().reset_index(name='First_purchase')
data_2
```

```python
#merging the first year of purchase data into oringinal table
data=pd.merge(data,data_2,left_on=['Customer ID'],right_on=['Customer ID'])
data
```

```python
#merging the first year of purchase data into oringinal table
data=pd.merge(data,data_2,left_on=['Customer ID'],right_on=['Customer ID'])
data
```

```python
#merging the two tables
data_f=pd.merge(data,data_2,left_on=['Customer ID'],right_on=['Customer ID'])
data_f
```

```python
#creating unique customer IDS
customer_ID=data_f['Customer ID'].unique()
#extracting the years of purchasing from data
years=range(data_f['order_year'].min(),data_f['order_year'].max()+1)
years
```

```python
#combining both s.t a combination of unique customer IDs and years of order are made
import itertools
data_l=list(itertools.product(customer_ID,years))
data_l=pd.DataFrame(data_l,columns=['Customer ID','year_of_order'])
data_l
```

```python
#commbined the data_l table and data_f table into data_n
data_n=pd.merge(data_l,data_f,how='left',left_on=['Customer ID','year_of_order'],right_on=['Customer ID','order_year'])
```

```python
#Create a field to flag these new rows, making it clear whether a customer placed an order in that year or not
data_n['Purchased']=np.where(data_n['Quantity']>0,1,0)
data_n.head(10)
```

```python
data_n['last_year_purch?']=data_n.groupby(['Customer ID'])['Purchased'].transform(lambda x:x.shift())
data_n
```

```python
from numpy import where
data_n['Customer Classification']=where(data_n['last_year_purch?'].isna(),'New',
                                  where((data_n['Purchased']==1) & (data_n['last_year_purch?']==1),'Consistent',
                                  where((data_n['Purchased']==1) & (data_n['last_year_purch?']==0),'Returning','Sleeping')))
```

```python
data_n
```

```python
#drop the first_purchase coloum from data_n since it contains NaN value
data_n=data_n.drop('First_purchase',axis=1)
data_n.info()
```

```python
data_m=pd.merge(data,data_n,how='left',left_on=['Customer ID'],right_on=['Customer ID'])
data_m.info()
```
